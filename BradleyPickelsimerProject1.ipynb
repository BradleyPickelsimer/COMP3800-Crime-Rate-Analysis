1a. Load the data (It should have at least 100 rows.) and get familiar with it. Describe the data source and include its link.

```python
import pandas as pd
crime_df = pd.read_csv('crimedata.csv')
print(crime_df)
```

The data source is a csv file found on kaggle that describes different crimes in different cities in the United Staes. The link to the dataset is: https://www.kaggle.com/datasets/michaelbryantds/crimedata

1b. List out three questions you want to investigate from this dataset. These questions could be a trend, a comparison between different groups of a categorical feature, a relationship between two variables, or a distribution of a numerical variable.

1b1: What are the 5 states with the highest crime rates?
1b2: Does unemployment rates have any correlation to crime rates?
1b3: Does income have any correlation to crime rates?

```python
cols = crime_df.columns
remove = cols[37:129]
crime_df = crime_df.drop(remove, axis = 1)
remove = cols[25:36]
crime_df = crime_df.drop(remove, axis = 1)
remove = cols[2:4]
crime_df = crime_df.drop(remove, axis = 1)
remove = cols[5:24]
crime_df = crime_df.drop(remove, axis = 1)
remove = ['murdPerPop','rapesPerPop','robbbPerPop','assaultPerPop','burglPerPop','larcPerPop','autoTheftPerPop','arsonsPerPop','ViolentCrimesPerPop','nonViolPerPop']
crime_df = crime_df.drop(remove, axis = 1)
print(crime_df)
```

```python
print(crime_df.isnull().sum())
```

```python
crime_df = crime_df.dropna(axis = 0)
print(crime_df.isnull().sum())
```

```python
crime_df = crime_df.sort_values(by = ['state'])
print(crime_df)
```

```python
crime_df['TotalCrimes'] = crime_df['rapes'] + crime_df['robberies'] + crime_df['assaults'] + crime_df['burglaries'] + crime_df['larcenies'] + crime_df['autoTheft'] + crime_df['arsons']
crime_df['NumberOfUnemployed'] = (crime_df['PctUnemployed'] / 100) * crime_df['population']
crime_df = crime_df.round({'NumberOfUnemployed' : 0})
print(crime_df)
```

3.1A) Which 5 states have the highest crime rate?

```python
crime1_df = crime_df.groupby(by = ['state']).sum()
print(crime1_df)
```

```python
crime1_df['CrimeRate'] = (crime1_df['TotalCrimes'] / crime1_df['population']) * 100
print(crime1_df)
```

```python
crime1_df = crime1_df.sort_values(by = ['CrimeRate'], ascending = False)
crime1_df = crime1_df.reset_index()
print(crime1_df)
```

3.1B) Create a graph to vizualize the result.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
plt.barh(crime1_df.iloc[0:11,0],crime1_df.iloc[0:11,13])
plt.xticks(np.arange(0, 15, 1))
plt.gca().invert_yaxis()
```

3.1C) Interpret the result. Explain your findings.

The top 5 states with the highest crime rate is Nevada, Maryland, Florida, Georgia, and Washington, D.C.

The results were found by grouping all of the cities together by state and adding up all of the columns to get 1 index to represent each state. The total number of cries was found by adding together all of the different types of crimes reported. To find the crime rate, the total number of crimes was divided by the total population for each state.

3.2A)  Does unemployment rates have any correlation to crime rates?

```python
cols = crime1_df.columns
remove = cols[1:11]
crime2_df = crime1_df.drop(remove, axis = 1)
crime2_df['UnemployedCrimeRate'] = crime2_df['TotalCrimes'] / crime2_df['NumberOfUnemployed']
print(crime2_df)
```

3.2B) Create a graph to vizualize the result.

```python
y = crime2_df['CrimeRate']
x = crime2_df['UnemployedCrimeRate']
plt.scatter(x, y, marker = 'o', color = 'c')
plt.xlabel('Unemployment')
plt.ylabel('Crime Rate')
```

3.2C) Interpret the result. Explain your findings.

For crime rate and unemployement, the scatterplot appears to show a moderatly strong, positive relationship. As unemployment increases, the crime rate also tends to increase.
